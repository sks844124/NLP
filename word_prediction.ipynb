```python
def get_cluster_LastWord(model, tokenizer, max_length, seed_text, n_words):
  in_text = seed_text
  for _ in range(n_words):
    encoded = tokenizer.texts_to_sequences([in_text])[0]
    encoded = pad_sequences([encoded], maxlen=max_length, padding='pre')
    y2,yi = model.predict(encoded, verbose=0)
    y2_max = np.argmax(y2, axis = 1)
    y2_max = int(y2_max)
    yi_max = np.argmax(yi, axis = 1)
    yi_max = int(yi_max)
    out_word = ''
    for word, index in tokenizer.word_index.items():
      if index == yi_max:
        out_word = word
        break
    y2_max = int(y2_max)    
    in_text += ' ' + out_word + ', cluster no = ' + str(y2_max) + ', word index of cluster no = ' + str(index)
  return in_text
```
